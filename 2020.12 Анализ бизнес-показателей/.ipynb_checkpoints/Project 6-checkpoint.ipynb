## Шаг 2. Построим отчёты и посчитаем метрики

### 2.1. Продукт.

#### 2.1.1. Определим сколько людей пользуются в день (DAU), неделю (WAU), месяц (MAU), а также регулярность использования сервиса (sticky factor).

Добавим в таблицу visits столбы с днем/неделей/месяцем сессии.

```python
visits['date'] = visits['start_ts'].dt.date
visits['week'] = visits['start_ts'].dt.strftime('%Y-%W')
visits['month'] = visits['start_ts'].dt.strftime('%Y-%m')
```

Определим DAU, WAU, MAU:

```python
dau = visits.pivot_table(index='date', values='user_id', aggfunc='nunique')['user_id'].mean()
wau = visits.pivot_table(index='week', values='user_id', aggfunc='nunique')['user_id'].mean()
mau = visits.pivot_table(index='month', values='user_id', aggfunc='nunique')['user_id'].mean()
print('DAU равно {:.2f}.'.format(dau))
print('WAU равно {:.2f}.'.format(wau))
print('MAU равно {:.2f}.'.format(mau))
```

Определим sticky factor.

```python
sticky_factor = dau/mau*100
print('Sticky factor равен {:.2f}.'.format(sticky_factor))
```

#### 2.1.2. Определим сколько сессий в день.

```python
sessions_amount = visits.pivot_table(index='date', values='start_ts', aggfunc='count')['start_ts'].mean()
print('Среднее число сессий в день равно {:.2f}.'.format(sessions_amount))
sessions_per_user = sessions_amount/dau
print('Количество сессий в день на одного пользователя {:.2f}.'.format(sessions_per_user))
```

#### 2.1.4. Определим как часто люди возвращаются (retention rate).

```python
visits_per_user = visits.groupby('user_id').agg({'start_ts':'count'})

visits_per_user.hist(bins=9, range=(1,10), figsize=(15,4))
plt.title('Гистограмма количества сессий одного пользователя', fontsize = 14, weight = 'bold')
plt.xlabel('Количество сессий')
plt.ylabel('Частота')
```

```python
len(visits_per_user[visits_per_user['start_ts']==1])/len(visits_per_user)*100
```

Определили, что 77% пользователей воспользовались сервисом лишь однажды.

```python
first_session = visits.groupby(['user_id'])['date'].min()
first_session.name = 'first_session'
visits = visits.join(first_session, on='user_id', how='left')
```

```python
visits['first_session_month'] = visits['first_session'].astype('datetime64[M]')
visits['session_month'] = visits['date'].astype('datetime64[M]')
```

```python
visits['lifetime'] = visits['session_month'] - visits['first_session_month']
visits['lifetime'] = visits['lifetime'] / np.timedelta64(1,'M')
visits['lifetime'] = visits['lifetime'].round().astype('int')
```

```python
cohorts = (visits
.groupby(['first_session_month','lifetime'])
.agg({'user_id':'nunique'}).reset_index()
)
```

```python
cohort_users = cohorts[cohorts['lifetime'] == 0][['first_session_month','user_id']]
cohort_users = cohort_users.rename(columns = {'user_id': 'cohort_users'})
cohorts = cohorts.merge(cohort_users, on = 'first_session_month')
cohorts['retention'] = cohorts['user_id'] / cohorts['cohort_users']
```

```python
cohorts['first_session_month_str'] = cohorts['first_session_month'].dt.strftime('%Y-%m-%d')
```

```python
ret_rate = cohorts.query('lifetime != 0').pivot_table(index = 'first_session_month_str',
columns = 'lifetime',
values = 'retention',
aggfunc = 'sum')
```

```python
sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('Retention Rate', fontsize = 14, weight = 'bold')
fig = sns.heatmap(ret_rate,
annot = True, fmt = '.1%',
linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('Возраст')
plt.ylabel('Когорта')
plt.show(fig)
```

Видим негативную динамику по всем когортам, а также то, что у каждой последующей когорты retention rate в перый месяц уменьшается.

#### 2.2.4. Определим сколько денег приносят (LTV).

LTV (от англ. lifetime value) — «пожизненная» ценность клиента. Это — итоговая сумма денег, которую один клиент приносит в среднем со всех своих покупок. Определим ее.

```python
orders['first_order_month'] = orders['first_order_time'].astype('datetime64[M]')
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
```

Посчитаем количество новых покупателей (n_buyers) за каждый месяц:

```python
cohort_sizes = orders.groupby('first_order_month').agg({'user_id': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку.

```python
cohorts = orders.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
```

Добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц:

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
```

Найдем валовую прибыль, умножив выручку на маржинальность. Так как LTV — относительный показатель, расчитаем его по «возрастным» когортам, где в столбцах — не месяц заказа, а возраст (age) когорты.

```python
margin_rate = 1
    
report['gp'] = report['revenue'] * margin_rate
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
```

Наконец, найдём LTV. Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте. Для наглядности визуализируем когорты в виде сводной таблицы:

```python
report['ltv'] = report['gp'] / report['n_buyers']
    
ltv = report.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean').cumsum(axis=1)
```

Построим диаграмму.

```python
sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('LTV', fontsize = 14, weight = 'bold')
fig = sns.heatmap(ltv,
annot = True, fmt = '.2f',
linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('Возраст')
plt.ylabel('Когорта')
plt.show(fig)
```

В связи с тем, что 77% пользователей заходят на сервис лишь однажды, самые большие значения прироста LTV в каждой из когорт - в первый месяц. LTV первой когорты - 11.88 у.е.

Определим как меняются затраты на маркетинг по отдельным источникам в течении времени. 

```python
costs['dt_month'] = costs['dt'].astype('datetime64[M]')
costs_pivot3 = costs.pivot_table(columns='dt_month', values='costs', index='source_id', aggfunc='sum')

def costs_plot(source_id):
    data = costs_pivot3.loc[source_id]

    plt.plot(data.index, data.values)
    
plt.figure(figsize = (15, 4))

plt.grid(True)

source_ids = [1, 2, 3, 4, 5, 9, 10]

for id in source_ids:
    costs_plot(id)
    
plt.legend(['1', '2', '3', '4', '5', '9', '10'])

plt.title('График затрат на маркетинг различных источников по месяцам', fontsize = 14, weight = 'bold')
plt.xlabel('Месяцы')
plt.ylabel('Затраты, у.е.')
```

#### 2.3.2. Определим сколько стоило привлечение одного покупателя из каждого источника.

Объеденим таблицы visits и orders по user_id, с целью присвоения каждому пользователю источника, откуда он впервые перешел на сайт.

```python
combined = pd.merge(orders, visits, left_on='user_id', right_on='user_id')[['first_order_month','source_id', 'user_id']]
```

```python
cohorts_source = combined.pivot_table(index='first_order_month', columns='source_id', values='user_id', aggfunc='nunique')
```

```python
costs_pivot4 = costs.pivot_table(index='dt_month', columns='source_id', values='costs', aggfunc='sum')
```

Рассчитаем CAC каждой когорты по каждому источнику.

```python
cac = costs_pivot4/cohorts_source

sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('CAC', fontsize = 14, weight = 'bold')
fig = sns.heatmap(cac,
annot = True, fmt = '.2f',
linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('Идентификатор рекламного источника')
plt.ylabel('Когорта')
plt.show(fig)
```

Рассчитаем средний CAC каждой когорты без учета источников.

```python
cac_mean = cac.mean(axis=1) 
cac_mean.name = 'cac_mean'
cac_mean
```

#### 2.3.3. Определим на сколько окупились расходы (ROMI).

Для определения ROI разделим LTV когорты на соответствующий CAC.

```python
romi = ltv.div(cac_mean, axis=0)
sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('ROMI', fontsize = 14, weight = 'bold')
fig = sns.heatmap(romi.cumsum(axis=1),
annot = True, fmt = '.2f',
linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('Возраст')

plt.ylabel('Когорта')
plt.show(fig)
```

<div style="border:solid red 2px; padding: 20px"> <b>Что необходимо доработать</b><br>
<ul>
    <li> цифры <code>ROMI</code> изменятся, когда пересчитаешь <code>CAC</code>
    <li> и ты решил разделить таблицу <code>ltv</code> только на значение <code>CAC</code> для первой когорты, вот тут <code>ltv.div(cac_sum.iloc[0], axis='columns')</code> - это не верно, для каждой когорты будет своё значение <code>CAC</code></ul></div><br>

<div class="alert alert-info" style="border:solid blue 2px; padding: 20px"> <b>Комментарий студента: </b> Сделал, но в таком варианте кажется некорректный получается результат.</div>

<div style="border:solid red 2px; padding: 20px"> <b>Что необходимо доработать 2</b><br>
Да, действительно, в таком варианте получается не корректный результат, потому что при таком подходе меняется механика вычислений.<br>
Давай воспользуемся проверенным способом, возьмем таблицу <code>report</code>, добавим в нее <code>CAC</code>, вычислим ROMI и сформируем сводную таблицу (не забудь накопительный эффект, <code>cumsum</code>, как у <code>LTV</code>)</div><br

```python
report = report.join(cac_mean, on='first_order_month')
```

```python
report['romi'] = report['ltv']/report['cac_mean']
```

```python
romi = report.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='romi', 
        aggfunc='mean').cumsum(axis=1)
```

```python
sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('ROMI', fontsize = 14, weight = 'bold')
fig = sns.heatmap(romi,
    annot = True, fmt = '.2f',
    linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('Возраст')
plt.ylabel('Когорта')
plt.show(fig)
```

Видим, что большинство когорт окупаются уже на первый или второй месяц, но три когорты (01-2018, 02-2018, 04-2018) не окупились по текущий момент (06-2018).

Определим ROMI средней когорты.

```python
romi.mean(axis=0) 
```

Видим, что в среднем к концу 1го месяца когорта окупается.
