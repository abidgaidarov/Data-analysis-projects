### 5.1. –ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ

–í—ã–¥–≤–µ–Ω–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ, –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ –∏ –æ–Ω–∏ —Ä–∞–∑–ª–∏—á–Ω—ã.

```python
df2 = df2.dropna(subset=['user_score'])
```

```python
alpha = 0.05

xone_rating = df2.query("platform == 'XOne'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values
pc_rating = df2.query("platform == 'PC'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values

results = st.ttest_ind(
    xone_rating, 
    pc_rating, 
    equal_var = False)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if (results.pvalue < alpha):
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

–í—ã–¥–≤–µ–Ω–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ, –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ –∏ –æ–Ω–∏ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è.

```python
alpha = 0.05

action_rating = df2.query("genre == 'Action'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values
sports_rating = df2.query("genre == 'Sports'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values

results = st.ttest_ind(
    action_rating, 
    sports_rating, 
    equal_var = False)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if (results.pvalue < alpha):
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```
