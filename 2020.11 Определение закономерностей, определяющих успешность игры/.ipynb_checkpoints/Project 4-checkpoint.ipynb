```python
df1['user_score'] = pd.to_numeric(df1['user_score'], errors='coerce')
```

### 4.3. Определим для пользователя каждого региона влияет ли на покупки рейтинг ESRB

Определим какие есть значения в колонке с рейтингом ESRB:

```python
df2['rating'].unique()
```

Здесь:
- E - "Everyone" - для всех возрастов;
- K-A - "Kids to Adults" - для всех возрастов;
- EC - "Early childhood" - для всех возрастов; 
- E10+ - "Everyone 10+" - для возрастов 10+;
- T - "Teen" - для возрастов 13+;
- M - "Mature" - для возрастов 18+;
- RP - "Rating pending" - рейтинг еще не присвоен.

С целью определения влияния рейтинга ESRB на покупки, заменим буквенные значения на условный порядковый номер:
- E, K-A, EC - 0;
- EC-10+ - 1;
- T - 2;
- M - 3.
- RP - отбросим, так как рейтинг не известен.

```python
df3 = df2.copy()
df3 = df3[df3.rating!='RP'].dropna(subset=['rating'])
```

```python
df3['rating_class'] = df3['rating'].replace(['E', 'M', 'T', 'E10+', 'K-A', 'EC'], [0, 3, 2, 1, 0, 1])
```

Определим корреляцию:

```python
df3['na_sales'].corr(df3['rating_class'])
```

```python
df3['eu_sales'].corr(df3['rating_class'])
```

```python
df3['jp_sales'].corr(df3['rating_class'])
```

```python
df3['other_sales'].corr(df3['rating_class'])
```

Определили, что зависимость во всех регионах отсутствует.

## Шаг 5. Проверим гипотезы

### 5.1. Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Выдвенем нулевую гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, и альтернативную гипотезу, что и они различны.

```python
df2 = df2.dropna(subset=['user_score'])
```

```python
alpha = 0.05

xone_rating = df2.query("platform == 'XOne'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values
pc_rating = df2.query("platform == 'PC'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values

results = st.ttest_ind(
    xone_rating, 
    pc_rating, 
    equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Определили, что средние пользовательские рейтинги платформ Xbox One и PC различны.

### 5.2. Проверим гипотезу, что средние пользовательские рейтинги жанров Action и Sports разные

Выдвенем нулевую гипотезу, что средние пользовательские рейтинги жанров Action и Sports одинаковые, и альтернативную гипотезу, что и они отличаются.

```python
alpha = 0.05

action_rating = df2.query("genre == 'Action'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values
sports_rating = df2.query("genre == 'Sports'").pivot_table(index='name', values='user_score', aggfunc='mean')['user_score'].values

results = st.ttest_ind(
    action_rating, 
    sports_rating, 
    equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Определили, что средние пользовательские рейтинги жанров Action и Sports разные.
